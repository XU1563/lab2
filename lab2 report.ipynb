```python
from Bio import SeqIO
seqid=['hu.39','rh.50','rh.57','rh.51','rh.49','rh.53','rh.64','rh.61','rh.58']
store=[]
for seg in SeqIO.parse('seqs.aligned.fa','fasta'):
    if seg.name in seqid:
        store.append(seg.seq)
# print (store)

result=[]
selected_seq=store
for position in range (len(selected_seq[0])):
    result.append([0,0,0])
    for seq in selected_seq:
        if seq[position]=='A' or seq[position]=='T':
            result[position][0]+=1
        elif seq[position]=='G' or seq[position]=='C':
            result[position][1]+=1
        else:
            result[position][2]+=1
            
# print (result)
store=[]
for length in range(len(result)):
    store.append([0,0,0])
    AT=result[length][0]
    CG=result[length][1]
    NA=result[length][2]
    total=AT+CG+NA
    ATPP=AT/(total*1.0)
    CGPP=CG/(total*1.0)
    NAPP=NA/(total*1.0)
    store[length][0]=ATPP
    store[length][1]=CGPP
    store[length][2]=NAPP

import numpy as np
import matplotlib.pyplot as plt
# Values of each group

barat=[]
barcg=[]
barna=[]

for length in range(len(store)):
    barat.append(store[length][0])
    barcg.append(store[length][1])
    barna.append(store[length][2])
# print(barat)
import numpy as np
bars=np.add(barat,barcg)
r=list(range(400,800))      # choose the sequence from the range 400-800 to plot 

#Create brown bars
barWidth = 1
plt.figure(figsize=(100,20))
plt.bar(r, barat[400:800], color='red', edgecolor='white', width=barWidth)
plt.bar(r, barcg[400:800], bottom=barat[400:800], color='blue', edgecolor='white', width=barWidth)
plt.bar(r, barna[400:800], bottom=bars[400:800], color='yellow', edgecolor='white', width=barWidth)
plt.xlabel('Position in Sequence',fontsize=100)
plt.ylabel('%AT or %GC',fontsize=100)
plt.tick_params(labelsize=80)
plt.show()       
```

```python
# Find the length of each sequence in each cluster
from Bio import SeqIO
ids = [('hu.39','rh.50','rh.57','rh.51','rh.49','rh.53','rh.64','rh.61','rh.58'),('pi.2','pi.3','pi.1','rh.10','bb.1','bb.2','hu.17','hu.6','rh.2','rh.40','hu.67','hu.37','hu.40','hu.66','hu.41','rh.38','hu.42'),('cy.3','cy.6','cy.4','cy.5','rh.13','rh.35','rh.36','rh.37','cy.2','rh.54', 'rh.55', 'rh.48', 'rh.62', 'hu.43', 'hu.48', 'hu.44', 'hu.46'),('hu.14', 'hu.31', 'hu.32','rh.43')]
num_clusters=len(ids)
seqs = {k: [] for k in range(0,num_clusters)} 

# tested for the aligned file 
for seg in SeqIO.parse('seqs.fa','fasta'): 
    for cluster in range(0,num_clusters):
          if seg.name in ids[cluster]:
                seqs[cluster].append(len(seg.seq))

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
box_plot_data=[]
for position in range (num_clusters):
    box_plot_data.append(seqs[position])
    
plt.boxplot(box_plot_data,labels=["cluster1","cluster2","cluster3","cluster4"])
plt.ylabel("Sequence Length")
ax = plt.axes() 
ax.grid(zorder=0)
bp = ax.boxplot(box_plot_data)
for box in bp['boxes']:
    # change outline color
    box.set(color='y', linewidth=2)
plt.show()
```
